**Here are the parameters to experiment with:**

- affix_length: we determines the length of affixes to consider. It can be a positive or negative integer. a negative value means suffixes will be considered, while a positive value means prefixes will be considered. example, -3 means considering suffixes of length 3, while 3 means considering prefixes of length 3.
- min_stem_length: we specifies the minimum stem length for a word to be considered for tagging. this parameter helps avoid tagging very short words where affixes might not be reliable indicators of the part of speech.

**By experimenting with different settings for affix_length and min_stem_length, we can observe how these parameters affect the performance of the affix tagger. Here are some observations we might make:**

 Affix Length:

- longer affix lengths might capture more information but could also lead to capturing noise.
- shorter affix lengths might generalize better but may miss some important patterns in the data.


Minimum Stem Length:

- increasing this value might improve accuracy by excluding very short words that may not have reliable affix patterns.
- also if we might try setting it too high, lead to excluding too many words, reducing coverage and potentially missing important patterns.
